```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

# Create a scatter plot
rel_plot = sns.relplot(
    x="Category",
    y="Installs",
    hue="Content Rating",
    data=dataset,
    kind="scatter",
    height=6,
    aspect=2,
)
# Rotate x-axis labels for better visibility
rel_plot.set_xticklabels(rotation=90)

# Show the plot
plt.show()
```

```python
plt.hist(dataset["Installs"]);
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(8,6))
plt.xticks(rotation=90)
sns.countplot(x="Rating",data=dataset)
plt.title("Rating")
plt.show()
```

```python
fig=plt.figure(figsize=(10,6))
plt.xticks(rotation=90)
sns.lineplot(x="Rating",y="Category",data=dataset)
```

```python
category_counts=dataset["Category"].value_counts()
category_counts.plot(kind="bar",color="green")
plt.xlabel("Category")
plt.ylabel("Number of Apps ")
plt.title("Number of Apps in each Category")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.hist(dataset["Rating"],bins=20,color="skyblue")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("App Rating Distribution")
plt.show()
```

```python
dataset["Size"] = dataset["Size"].replace("Varies with device", pd.NA)
dataset["Size"] = dataset["Size"].astype(str).str.extract('([\d\.]+)').astype(float)
average_size_by_category = dataset.groupby('Category')['Size'].mean()
average_size_by_category = average_size_by_category.sort_values(ascending=False)
plt.figure(figsize=(12, 12))
plt.pie(average_size_by_category, labels=average_size_by_category.index, autopct="%1.1f%%", startangle=140)
plt.title("Average App Size Distribution by Category")
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x='Type', y='Installs', data=dataset, palette='Set2')
plt.yscale('log')
plt.title('Distribution of Installs by Type')
plt.xlabel('Type')
plt.ylabel('Installs (Log Scale)')
plt.show()
```

```python
free_installs = dataset[dataset['Type'] == 'Free']['Installs']
paid_installs = dataset[dataset['Type'] == 'Paid']['Installs']
plt.figure(figsize=(10, 6))
sns.histplot(free_installs, bins=30, kde=True, label='Free', color='blue', alpha=0.7)
sns.histplot(paid_installs, bins=30, kde=True, label='Paid', color='orange', alpha=0.7)
plt.xscale('log')
plt.xlabel('Installs (Log Scale)')
plt.ylabel('Frequency')
plt.title('Distribution of Installs by Type')
plt.legend()
plt.show()
```

```python
dataset = dataset.sort_values(by="Installs", ascending=False)
top_50_apps = dataset.head(50)
plt.figure(figsize=(14, 10))
plt.barh(top_50_apps["App"], top_50_apps["Installs"], color="skyblue")
plt.xlabel("Installs")
plt.title("Top 50 Most Installed Apps (String Format)")
plt.tight_layout()
plt.show()
```

```python
dataset["Price"] = pd.to_numeric(dataset["Price"].str.replace('[^0-9.]', '', regex=True), errors="coerce")
dataset = dataset.sort_values(by="Price", ascending=False)
top_50_expensive_apps = dataset.head(50)
plt.figure(figsize=(14, 10))
plt.scatter(top_50_expensive_apps["App"], top_50_expensive_apps["Price"], color="red")
plt.xlabel("App")
plt.ylabel("Price")
plt.title("Top 50 Most Expensive Apps")
plt.xticks(rotation=90, ha="center")
plt.yscale("log")
plt.tight_layout()
plt.show()
```

```python
dataset['Last Updated'] = pd.to_datetime(dataset['Last Updated'], errors='coerce')
dataset['YearMonth'] = dataset['Last Updated'].dt.to_period("M")
apps_by_last_updated = dataset['YearMonth'].value_counts().sort_index()
plt.figure(figsize=(12, 8))
plt.plot(apps_by_last_updated.index.astype(str), apps_by_last_updated.values, marker="d", linestyle="--", color="red")
plt.title("Number of Apps by Last Updated Date")
plt.xlabel("Year-Month")
plt.ylabel("Number of Apps")
plt.xticks(rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(y="Android Ver",data=dataset,order=dataset["Android Ver"].value_counts().index,palette="viridis")
plt.title("Distribution of Android Versions")
plt.xlabel("Number of Apps")
plt.ylabel("Android Version")
plt.tight_layout()
plt.show()
```

```python
dataset["Installs"] = dataset["Installs"].apply(lambda x: pd.to_numeric(str(x).replace('[^0-9]', ''), errors='coerce'))
plt.figure(figsize=(14, 8))
sns.boxplot(x="Installs", y="Category", data=dataset, palette="viridis")
plt.xscale("log")
plt.title("Distribution of Installs by Category")
plt.xlabel("Number of Installs (log scale)")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```
